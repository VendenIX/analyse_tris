 Résultats des performances des algorithmes de tri selon désordre
 ====================================================

Ici nous allons présenter les différents résultats de nos expérimentations en terme de temps. ( Eventuellement écrire les conditions de test des algos genre vite fait) Cette présentation aura pour dessein de comparer les différents algorithmes dans de multiples conditions, de ces comparaisons nous obtiendrons une idée globale de l'efficacité de ceux-ci pour notre analyse.

Dans un premier temps, nous avons décidé de découper nos résultats par type d'objets à trier, ainsi nous verrons, dans l'odre les résultats sur le tri des int, tableau de int et String.

Dans un second temps, nous diviserons ces recherches par type de désordre.

```python
import pandas as pd
import numpy as np
import feather as fe
import matplotlib.pyplot as plt
import time
```

```python
df_size = 1000

BubbleInt = pd.read_csv('result_data/BubbleSort.csv',sep=',')# On récupère les données correspondantes
print(BubbleInt)
ShellInt = pd.read_csv('result_data/ShellSort.csv',sep=',')
GnomeInt = pd.read_csv('result_data/GnomeSort.csv',sep=',')
SelectionInt=pd.read_csv('result_data/SelectionSort.csv',sep=',')
BingoInt=pd.read_csv('result_data/BingoSort.csv',sep=',')
PancakeInt=pd.read_csv('result_data/PancakeSort.csv',sep=',')
CombInt=pd.read_csv('result_data/CombSort.csv',sep=',')
HeapInt=pd.read_csv('result_data/HeapSort.csv',sep=',')
InsertionInt=pd.read_csv('result_data/InsertionSort.csv',sep=',')
QuickInt=pd.read_csv('result_data/QuickSort.csv',sep=',')

def AttributeColumns(DF):
    DF.columns = ["NumberElements","NombreEchangeRd",
                     "NombreComparaisonRd","TempsMSRd","NombreEchange1_2",
                     "NombreComparaison1_2","TempsMS1_2","NombreEchangeStart25",
                     "NombreComparaisonStart25","TempsMSStart25",
                     "NombreEchangeStart50","NombreComparaisonStart50",
                     "TempsMSStart50","NombreEchangeEnd25",
                     "NombreComparaisonEnd25","TempsMSEnd25",
                     "NombreEchangeEnd50","NombreComparaisonEnd50",
                     "TempsMSEnd50","NombreEchangeSorted",
                     "NombreComparaisonSorted","TempsMSSorted"]
AttributeColumns(BubbleInt)
AttributeColumns(ShellInt)
AttributeColumns(GnomeInt)
AttributeColumns(SelectionInt)
AttributeColumns(BingoInt)
AttributeColumns(PancakeInt)
AttributeColumns(CombInt)
AttributeColumns(InsertionInt)
AttributeColumns(QuickInt)
AttributeColumns(HeapInt)
    
def Writedf(nom):
    df_size = 1000
    df = pd.DataFrame({
    'Elements':nom.loc[:,'NumberElements'],
    'VitesseSorted' : nom.loc[:,'TempsMSSorted'],
    'VitesseRandom' : nom.loc[:,'TempsMSRd'],
    'VitesseSorted1_2': nom.loc[:,'TempsMS1_2'],
    'VitesseMixedStart25': nom.loc[:,'TempsMSStart25'],
    'VitesseMixedEnd25': nom.loc[:,'TempsMSEnd25'],
    'VitesseMixedStart50': nom.loc[:,'TempsMSStart50'],
    'VitesseMixedEnd50': nom.loc[:,'TempsMSEnd50'],
    'EchangeSorted' : nom.loc[:,'NombreEchangeSorted'],
    'EchangeRandom' : nom.loc[:,'NombreEchangeRd'],
    'EchangeSorted1_2': nom.loc[:,'NombreEchange1_2'],
    'EchangeMixedStart25': nom.loc[:,'NombreEchangeStart25'],
    'EchangeMixedEnd25': nom.loc[:,'NombreEchangeEnd25'],
    'EchangeMixedStart50': nom.loc[:,'NombreEchangeStart50'],
    'EchangeMixedEnd50': nom.loc[:,'NombreEchangeEnd50'],
    'ComparaisonSorted' : nom.loc[:,'NombreComparaisonSorted'],
    'ComparaisonRandom' : nom.loc[:,'NombreComparaisonRd'],
    'ComparaisonSorted1_2': nom.loc[:,'NombreComparaison1_2'],
    'ComparaisonMixedStart25': nom.loc[:,'NombreComparaisonStart25'],
    'ComparaisonMixedEnd25': nom.loc[:,'NombreComparaisonEnd25'],
    'ComparaisonMixedStart50': nom.loc[:,'NombreComparaisonStart50'],
    'ComparaisonMixedEnd50': nom.loc[:,'NombreComparaisonEnd50']
    })
    return df

BubbleDF=Writedf(BubbleInt)
ShellDF=Writedf(ShellInt)
SelectionDF=Writedf(SelectionInt)
GnomeDF=Writedf(GnomeInt)
BingoDF=Writedf(BingoInt)
PancakeDF=Writedf(PancakeInt)
CombDF=Writedf(CombInt)
HeapDF=Writedf(HeapInt)
InsertionDF=Writedf(InsertionInt)
QuickDF=Writedf(QuickInt)

```
